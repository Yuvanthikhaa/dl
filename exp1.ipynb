```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("D:/dl exps/lab programs/dl ex 1/diabetes.csv")
```

```python
data
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1].astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(12, input_dim=8, activation ='relu'))#first hidden layer
model.add(Dense(8, activation='relu'))#hidden layer same activation funation to be used in input adn hidden layer
model.add(Dense(1,activation='sigmoid'))#output layer binary classifiaction so sigmoid is used
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=150,batch_size=10)
```

```python
y_pred = model.predict(x_test).astype(int)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
